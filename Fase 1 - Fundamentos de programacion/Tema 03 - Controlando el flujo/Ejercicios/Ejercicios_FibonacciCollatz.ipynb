### Serie Fibonacci

```python
num_repeticiones = int(input ('Introduce numero repeticiones'))
repeticion = 0 
a, b = 0,1
while repeticion < num_repeticiones:
    print(a, end=', ')
    a, b = b, a+b
    repeticion+=1
```

### Conjetura Collatz

```python
num=int(input ('Introduce un numero'))
while (num!=1):
    print(num)
    if num%2==0:
        num = num/2
    else:
        num = 1+num*3

print(num)
```
