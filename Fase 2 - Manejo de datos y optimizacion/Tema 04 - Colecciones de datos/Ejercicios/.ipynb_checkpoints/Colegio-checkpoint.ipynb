```python
clase1 = {}
clase2 = {}
clase3 = {}

colegio = [clase1,clase2,clase3]
```

```python
opcion = 0
while opcion!=5:
    
    print("""
        1. Insertar Alumno\n
        2. Borrar Alumno
        3. Mostrar alumnos
        4. Mostrar alumno con nota mas alta
        5. Salir\n
    """)
    opcion=int(input('Elige una opcion'))

    if opcion == 1:
        dni=input('DNI ')
        apellidos=input('Apellidos ')
        nombre=input('Nombre ')
        nota=float(input('Nota '))
        
        email_correcto = False
        while not email_correcto:
            email=input('email ').strip()
            if (email.count('@')==1 and email.find(' ')==-1 and  (email.find('.')>email.find('@')+1)  ):
                email_correcto = True
            else:
                print("Email incorrecto, vuelve a introducirlo.")
        
        num_clase = int(input('En que clase quieres meter al alumno (1-3)'))
        
        #diccionario_clase = colegio[num_clase-1]
        #diccionario_clase[dni]=[nombre,apellidos,nota,email]
        
        colegio[num_clase-1][dni]=[nombre,apellidos,nota,email]
        
    elif opcion ==2:
        
        dni=input('DNI ')
        dni_eliminado = False
        for aula in colegio:
            if dni in aula:
                del(aula[dni])
                dni_eliminado = True
                break
        
        if dni_eliminado:
            print("Dni ", dni, " eliminado ")
        else:
            print("Dni ", dni, " NO eliminado ")
            
    elif opcion == 3:
        #Primera forma de hacerlo
        num_aula_bis = 0
        for alumnos in colegio:
            print("Alumnos del aula ",num_aula_bis+1)
            num_aula_bis +=1
            for dni,datos in alumnos.items():
                print (dni + " "+ datos[0] + " " + datos[1])
        
        #Segunda forma de hacerlo
        for num_aula in range (len(colegio)):
            print("Alumnos del aula ",num_aula+1)
            for dni,datos in colegio[num_aula].items():
                print (dni + " "+ datos[0] + " " + datos[1] )
            
    elif opcion == 4:
        nota_maxima = 0
        dni_alumno_nm = ''#Este va a ser el dni del alumno que tenga la nota mas alta
        for alumnos in colegio:
            for dni,datos in alumnos.items():
                if datos[2] > nota_maxima:
                    nota_maxima = datos[2]
                    dni_alumno_nm = dni
         
        print ( "El alumno" , alumnos[dni_alumno_nm][0] ," tiene la nota mas alta ",nota_maxima)
        
print("Adios")
```
